```python
class MnistBags():
  def __init__(self):
    pass

  def to_bagged(self,data,size):

    iterator = iter(data)
    i = 0
    
    baggedData = []
    baggedLabels = []
    while True:
      
      bagSize = tf.random.uniform(shape = [],
                            minval=self.minBS,
                            maxval=self.maxBS+1,
                            dtype=tf.int32)
      
      
      Bag = []
      bagTarget = False
      i += bagSize

      if i >= size:
        break

      
      for _ in range(bagSize):
        image,target = iterator.get_next()

        if target == self.target:
          bagTarget = True 

        Bag.append(tf.cast(image,dtype=tf.float32)/255.)

      for _ in range(self.maxBS - bagSize):
        Bag.append(tf.zeros(shape = self.imageSize, dtype=tf.float32))
        

      baggedData.append(Bag)
      baggedLabels.append(tf.cast(bagTarget,dtype=tf.float32))

    return baggedData,baggedLabels

  def to_generator(self,bags,labels):
    

    def gen():
      
      for bag,label in zip(bags,labels):

        yield bag, label
          

    return tf.data.Dataset.from_generator(
        gen,
        output_types = (tf.float32,tf.float32),
        output_shapes = ((self.maxBS,*self.imageSize),())
    )


  def load_data(self):
    mnist,self.info = tfds.load(name="mnist",
                           split=["train","test"],
                           as_supervised = True,
                           with_info = True
    )
    self.trainSize = self.info.splits["train"].num_examples
    self.testSize = self.info.splits["test"].num_examples
    self.imageSize = self.info.features["image"].shape
    self.targetSize = self.info.features["label"].shape

    return mnist

  def fit(self,target,minBagSize=3,maxBagSize=7,
          trainBatch=50,testBatch=50,
          BaP=True):
    
    train,test = self.load_data()

    self.target = target
    self.minBS = minBagSize
    self.maxBS = maxBagSize

    trainBags, trainLabels = self.to_bagged(train,self.trainSize)
    testBags, testLabels = self.to_bagged(test,self.testSize)


    train = self.to_generator(trainBags,trainLabels)
    test = self.to_generator(testBags,testLabels)

    if BaP:
      return self.batch_and_prefetch(train,
                                test,
                                trainBatch,
                                testBatch)
      

    return train , test
  



  def batch_and_prefetch(self,train,test,batchTrain,batchTest):
    return train.repeat().batch(batch_size=batchTrain).prefetch(tf.data.AUTOTUNE), \
            test.repeat().batch(batch_size=batchTest).prefetch(tf.data.AUTOTUNE)
```

```python
from tensorflow import keras
from tensorflow.keras import layers,metrics,losses
```

```python
class MILattention(keras.Model):
  def __init__(self,batchSize,bagSize,imageSize):
    super().__init__()
    self.L =  512
    self.D = 128
    self.K = 1
    self.batchSize = batchSize
    self.bagSize = bagSize
    self.imageSize = imageSize
    
    self.features_extractor = self.get_features_extractor(*self.imageSize,self.L)
    self.attention_layer = self.get_attention_layer(self.L,self.D,self.K)
    self.classifier = self.get_classifier(self.L)
    self.loss_estimator = losses.BinaryCrossentropy()

  def summary(self):
    self.features_extractor.summary()
    self.attention_layer.summary()
    self.classifier.summary()

  def compile(self, optimizer, **kwargs):
    super().compile(**kwargs)
    self.optimizer = optimizer

    self.loss_tracker = metrics.Mean(name = "c_loss")
    self.accuracy_tracker = metrics.BinaryAccuracy(name = "c_acc")

    
    

  def get_features_extractor(self,height,length,channels,outputSize):
    return keras.Sequential(
        [
        keras.Input(shape = (height,length,channels)),
        layers.Conv2D(filters = 20, kernel_size=5, activation="relu"),
        layers.MaxPool2D(pool_size = (2,2), strides = 2),
        layers.Conv2D(filters = 50, kernel_size=5, activation="relu"),
        layers.MaxPool2D(pool_size= (2,2), strides = 2),
        layers.Flatten(),
        layers.Dense(outputSize,activation="relu")
        ],
        name = "features_extractor"
    )

  def get_attention_layer(self,inputSize,outputSize1,outputSize2):
    return keras.Sequential(
        [
        keras.Input(shape=(inputSize)),
        layers.Dense(outputSize1,activation="tanh"),
        layers.Dense(outputSize2)
        ],
        name = "attention_layer"
    )

  def get_classifier(self,inputSize):
    return keras.Sequential(
        [
        keras.Input(shape = (inputSize)),
        layers.Dense(1,activation="sigmoid")
        ],
        name = "classifier"
    )

  @property
  def metrics(self):
    return [
            self.loss_tracker,
            self.accuracy_tracker
    ]

  def forward_pass(self,bags,return_weights = False,batchSize = None):
      if not batchSize:
        batchSize = self.batchSize

    
      concBags = tf.reshape(bags,(batchSize*self.bagSize, *self.imageSize))

      features = self.features_extractor(concBags)

      attention_weights = self.attention_layer(features)

      attention_weights = tf.reshape(attention_weights,[batchSize,self.bagSize])

      softmax = layers.Softmax()

      attention_weights = tf.expand_dims(softmax(attention_weights), -2)

      features = tf.reshape(features, [batchSize,self.bagSize,self.L])

      pooling = tf.squeeze(tf.matmul(attention_weights,features))

      y_pred = self.classifier(pooling)

      if return_weights:
        return attention_weights, y_pred

      return y_pred


  def train_step(self,data):
    bags,labels = data



    with tf.GradientTape() as tape:

      y_pred = self.forward_pass(bags)

      loss = self.loss_estimator(labels,y_pred)
      
    learnable_weights = self.features_extractor.trainable_weights + \
                        self.attention_layer.trainable_weights + \
                        self.classifier.trainable_weights

    grads = tape.gradient(
        loss,
        learnable_weights
    )

    self.optimizer.apply_gradients(
        zip(
            grads,
            learnable_weights
        )
    )


    self.loss_tracker.update_state(loss)
    self.accuracy_tracker.update_state(labels,y_pred)

    return {m.name: m.result() for m in self.metrics}

  def test_step(self,data):

    bags,labels = data

    y_pred = self.forward_pass(bags)

    loss = self.loss_estimator(labels,y_pred)

    self.loss_tracker.update_state(loss)
    self.accuracy_tracker.update_state(labels,y_pred)

    return {m.name: m.result() for m in self.metrics}

  def call(self,inputs,Training=False):
 
    shape = inputs.shape
    attention_weights,labels = self.forward_pass(inputs,
                                                 return_weights = True,
                                                 batchSize=shape[0])


    attention_weights = tf.squeeze(attention_weights)

    fig,axs = plt.subplots(nrows=shape[0],ncols=shape[1],figsize=(8,8))
    fig.patch.set_facecolor('black')

    for j,(input,row) in enumerate(zip(inputs,axs)):

      color = ((1.,0.,0.) if labels[j]<0.5 else (0.,1.,0.))
      for i,(image,ax) in enumerate(zip(input,row)):


        tmpColor = color * attention_weights[j][i]
        ax.imshow(tf.squeeze(image),cmap="gray")
        ax.set_yticklabels([])
        ax.set_xticklabels([])
        ax.set_xticks([])
        ax.set_yticks([])

        for spine in ax.spines.values():

          spine.set_edgecolor(tmpColor)

        
          spine.set_linewidth(4)
          
        

    plt.show()
    return attention_weights,labels

```
